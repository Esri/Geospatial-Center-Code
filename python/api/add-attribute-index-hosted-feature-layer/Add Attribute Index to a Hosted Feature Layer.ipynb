## Add Attribute Index to a Hosted Feature Layer
This notebook will add an attribute index to a hosted feature layer

```python
from arcgis.gis import GIS
from arcgis.features import FeatureLayer
gis = GIS("home")
```

Get a reference to the item that has the feature layer you want

```python
item = gis.content.get("a6e330c275dd421a9a3dda6e8e546f3d")
item
```

Create a `FeatureLayer` using its url

```python
fl = FeatureLayer(item['layers'][0].url)
fl
```

Specify the attribute index details

```python
indexes = {
    "indexes": [
        {
            "name": "iso3_index",
            "fields": "iso3",
            "isAscending": True,
            "isUnique": False,
            "description": "iso3 country index"
        }
    ]
}
```

Add the index to the feature layer's definition

```python
res = fl.manager.add_to_definition(indexes)
res
```
